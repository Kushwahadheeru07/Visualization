```python
plt.figure() ##page
plt.subplot(1,2,1) ##Divide page into two parts 1=page ,2=two part ,1-axis=1
plt.plot(x,np.sin(x))

plt.subplot(1,2,2)
plt.plot(x,np.cos(x))

plt.show()
```

```python
##DIviding our page

plt.figure() 
plt.subplot(2,2,1) 
plt.plot(x,np.sin(x))

plt.subplot(2,2,2)
plt.plot(x,np.cos(x))

plt.subplot(2,2,3)
plt.plot(x,np.tan(x))

plt.subplot(2,2,4)
plt.plot(x,np.cos(x+1))

plt.show()
```

```python
##DIviding our page vertically

plt.figure() 
plt.subplot(2,1,1) 
plt.plot(x,np.sin(x))

plt.subplot(2,1,2)
plt.plot(x,np.cos(x))



plt.show()
```

```python
plt.plot(x,np.sin(x-0),color='blue')
plt.plot(x,np.sin(x-1),color='g')
plt.plot(x,np.sin(x-2),color=(1.0,0.5,0))  ##rgb color
```

```python
##changing limit of x - axix and y- plot
plt.plot(x,np.sin(x-0),color='blue')
plt.ylim(-1.1,1.5)
plt.xlim(2,14)
plt.show()
```

```python
##Both axes are equal(filling empty space)
plt.plot(x,np.sin(x-0),color='blue')
plt.ylim(-1.1,1.5)
plt.xlim(2,14)
plt.axis('equal')
plt.show()
```

```python
##See the number system
plt.plot(x,np.sin(x-0),color='blue')
plt.ylim(-1.1,1.5)
plt.xlim(2,14)
plt.axis('equal')
plt.grid()
plt.show()
```

```python
##Adding title of the plot
##Adding label of x ,y
plt.plot(x,np.sin(x-0),color='blue')
plt.ylim(-1.1,1.5)
plt.xlim(2,14)
plt.title('this is sin plot before real datset execution to understan basic syntax')
plt.axis('equal')
plt.xlabel('x-axis')
plt.ylabel('y label')
plt.grid()
plt.show()
```

```python
plt.plot(x,np.sin(x-0),color='blue',label='Sinx')
plt.plot(x,np.cos(x-0),color='red',label='Cosx')
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.title('this is sin plot before real datset execution to understan basic syntax')
#plt.axis('equal')
plt.xlabel('x-axis')
plt.ylabel('y label')
plt.grid()
plt.legend()
plt.show()
```
